## Gather more translated articles

### English Livestock Scrape

```python
response = requests.get('http://www.fao.org/livestock-environment/en/')
soup = BeautifulSoup(response.text)
```

```python
article = ''
for item in soup.find_all('p'):
    article = article + ' ' + item.text
```

```python
with open("livestock_en.txt", "w") as text_file:
    text_file.write(article)
```

### Spanish Livestock Scrape and Translate

```python
response = requests.get('http://www.fao.org/livestock-environment/es/')
soup = BeautifulSoup(response.text)
```

```python
article = ''
for item in soup.find_all('p'):
    article = article + ' ' + item.text
```
